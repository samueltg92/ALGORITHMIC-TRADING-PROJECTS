```python
import pandas as pd
```

LOAD DATASETS

```python
df2019_2022 = pd.read_csv('1. EURUSD_Candlestick_15_M_BID_04.04.2019-04.04.2022.csv')
df2016_2019 = pd.read_csv('2. EURUSD_Candlestick_15_M_BID_05.04.2016-05.04.2019.csv')
df2013_2016 = pd.read_csv('3. EURUSD_Candlestick_15_M_BID_06.04.2013-06.04.2016.csv')
df2010_2013 = pd.read_csv('4. EURUSD_Candlestick_15_M_BID_07.04.2010-07.04.2013.csv')
df2007_2010 = pd.read_csv('5. EURUSD_Candlestick_15_M_BID_08.04.2007-08.04.2010.csv')
df2004_2007 = pd.read_csv('6. EURUSD_Candlestick_15_M_BID_09.04.2004-09.04.2007.csv')
df2004_2007
```

MERGE FUNCTION

```python
def fun_merge(df1, df2):
    return pd.merge(df1, df2, how='outer')
```

```python
df_merge2004_2010 = fun_merge(df2004_2007, df2007_2010)
df_merge2004_2010
```

```python
df_merge2004_2013 = fun_merge(df_merge2004_2010, df2010_2013)
df_merge2004_2013
```

```python
df_merge2004_2016 = fun_merge(df_merge2004_2013, df2013_2016)
df_merge2004_2016
```

```python
df_merge2004_2019 = fun_merge(df_merge2004_2016, df2016_2019)
df_merge2004_2019
```

```python
EURUSD_15m_2004_2022 = fun_merge(df_merge2004_2019, df2019_2022)
EURUSD_15m_2004_2022
```

EXPORTING TO CSV

```python
EURUSD_15m_2004_2022.to_csv('../data_sets/EURUSD_Candlestick_15_M_BID_09.04.2004-04.04.2022.csv')
```
