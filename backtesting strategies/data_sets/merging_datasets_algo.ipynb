```python
import pandas as pd
```

LOAD DATASETS

```python
df2019_2022 = pd.read_csv('1. AUDUSD_Candlestick_15_M_BID_16.04.2019-16.04.2022.csv')
df2016_2019 = pd.read_csv('2. AUDUSD_Candlestick_15_M_BID_17.04.2016-17.04.2019.csv')
df2013_2016 = pd.read_csv('3. AUDUSD_Candlestick_15_M_BID_18.04.2013-18.04.2016.csv')
df2010_2013 = pd.read_csv('4. AUDUSD_Candlestick_15_M_BID_19.04.2010-19.04.2013.csv')
df2007_2010 = pd.read_csv('5. AUDUSD_Candlestick_15_M_BID_20.04.2007-20.04.2010.csv')
df2004_2007 = pd.read_csv('6. AUDUSD_Candlestick_15_M_BID_21.04.2004-21.04.2007.csv')
df2004_2007
```

MERGE FUNCTION

```python
def fun_merge(df1, df2):
    return pd.merge(df1, df2, how='outer')
```

```python
df_merge2004_2010 = fun_merge(df2004_2007, df2007_2010)
df_merge2004_2010
```

```python
df_merge2004_2013 = fun_merge(df_merge2004_2010, df2010_2013)
df_merge2004_2013
```

```python
df_merge2004_2016 = fun_merge(df_merge2004_2013, df2013_2016)
df_merge2004_2016
```

```python
df_merge2004_2019 = fun_merge(df_merge2004_2016, df2016_2019)
df_merge2004_2019
```

```python
AUDUSD_15m_2004_2022 = fun_merge(df_merge2004_2019, df2019_2022)
AUDUSD_15m_2004_2022
```

EXPORTING TO CSV

```python
AUDUSD_15m_2004_2022.to_csv('../data_sets/AUDUSD_Candlestick_15_M_BID_21.04.2004-16.04.2022.csv')
```
